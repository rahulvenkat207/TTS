```python
import torch
from parler_tts import ParlerTTSForConditionalGeneration
from transformers import AutoTokenizer
import  soundfile as sf
```

```python
device = "cudo:0" if torch.cuda.is_available() else "cpu"
```

```python
model = ParlerTTSForConditionalGeneration.from_pretrained("parler-tts/parler-tts-mini-v1").to(device)
tokenizer = AutoTokenizer.from_pretrained("parler-tts/parler-tts-mini-v1")
```

```python
prompt = "Kohli Kohli around the MCG, Is Virat going for two? No, Virat says I will take the strike. He still has a lot to do. Kohli goes down the ground, Kohli goes out of the ground. This is a sensational hit from Virat Kohli. 22 needed off 7. Kohli goes over fine leg. He might have cleared the fine leg, he has!!! What a way to end that Haris Rauf over that was looking so good. Two sixes. India now needs 16 off 6."
description = "Jon's voice is monotone yet slightly fast in delivery, with a very close recording that almost has no background noise."

```

```python
input_ids = tokenizer(description, return_tensors="pt").input_ids.to(device)
prompt_input_ids = tokenizer(prompt, return_tensors="pt").input_ids.to(device)
```

```python
generation = model.generate(input_ids=input_ids, prompt_input_ids=prompt_input_ids)
audio_arr = generation.cpu().numpy().squeeze()
sf.write("parler_tts_out.wav", audio_arr, model.config.sampling_rate)
```
