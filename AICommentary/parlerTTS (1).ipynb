```python
import torch
from parler_tts import ParlerTTSForConditionalGeneration
from transformers import AutoTokenizer
import  soundfile as sf
```

```python
device = "cudo:0" if torch.cuda.is_available() else "cpu"
```

```python
model = ParlerTTSForConditionalGeneration.from_pretrained("parler-tts/parler-tts-mini-v1").to(device)
tokenizer = AutoTokenizer.from_pretrained("parler-tts/parler-tts-mini-v1")
```

```python
prompt = "The tension is building, the crowd is on their feet, the noise is deafening. Here comes the final ball, four runs needed to win the World Cup...and Dhoni does it! He's smashed it for six! India are world champions! This is an historic moment, a moment of pure ecstasy. The cheers are deafening, the Indian players are dancing in delight. They've achieved something truly remarkable. This is their day, their moment! The world is watching, and India are standing tall. What a day for cricket, what a day for India!"
description = "Jon's voice is monotone yet slightly slow in delivery and it should be emotional, with a very close recording that almost has no background noise."

```

```python
input_ids = tokenizer(description, return_tensors="pt").input_ids.to(device)
prompt_input_ids = tokenizer(prompt, return_tensors="pt").input_ids.to(device)
```

```python
generation = model.generate(input_ids=input_ids, prompt_input_ids=prompt_input_ids)
audio_arr = generation.cpu().numpy().squeeze()
sf.write("parler_tts_out.wav", audio_arr, model.config.sampling_rate)
```
